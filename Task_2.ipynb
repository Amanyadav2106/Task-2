```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('train_titan.csv')
```

```python
first_class = (data['Pclass']==1).sum()
second_class =(data['Pclass']==2).sum()
third_class =(data['Pclass']==3).sum()

plt.figure(figsize=(5,5))
color=['DeepSkyBlue', 'DarkTurquoise','Aqua']

plt.pie([first_class, second_class, third_class], labels=['First Class', 'Second Class', 'Third Class'], colors=color, autopct='%1.1f%%', startangle=90)

plt.show()
```

```python
count = data['Survived'].value_counts()

plt.figure(figsize=(5,5))
color=['pink', 'MistyRose']

label=['Not Survived', 'Survived']
plt.pie(count, labels=label, startangle=90, autopct='%1.1f%%', colors=color)
plt.title('Survived vs Not survived')

plt.show()
```

```python
sns.countplot(data, x = 'Survived')

plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survived Count')

plt.show()
```

```python
sns.countplot(data, x = 'Pclass')

Pclass_counts = data['Pclass'].value_counts()

sns.barplot(x=Pclass_counts.index, y=Pclass_counts.values, color='OrangeRed')

plt.xlabel('Pclass')
plt.ylabel('Count')
sns.set_style('dark')
plt.title('Distribution of Pclass')

plt.show()
```
